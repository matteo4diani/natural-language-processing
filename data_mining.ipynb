```python
from pprint import pprint
from bs4 import BeautifulSoup

def flat_map(f, xs):
    ys = []
    for x in xs:
        ys.extend(f(x))
    return ys

# Path to your Akoma Ntoso XML file
file_path = "/content/costituzione.xml"
from lxml import etree

root = None
soup = None

with open(file_path) as file:
  parser = etree.XMLParser(recover=True)
  root = etree.fromstring(file.read(), parser)

with open(file_path) as file:
  soup = BeautifulSoup(file.read(), 'html.parser')

#print(soup.prettify())
capi = soup.find_all('capo')
import bs4
for capo in capi:
  capo: bs4.element.Tag = capo
  print(''.join(map(lambda s: s.text, capo.find_all('h:p'))))
```
